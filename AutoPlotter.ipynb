# Import libraries

```python
from autoplotter import run_app
import pandas as pd
```

# Load data

```python
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/nba-player-stats-2019.csv')
# df = dataframe
```

```python
df
```

# run_app

```python
run_app(df, mode = 'inline') # Display the plot in Google Collab or Jupyter Notebook
# Data Preview: you could have aprooximately 15 rows of data per page and you could filter out the data you want.
# Plots: you could choose from seaborn to plotly to present the plots through bar chart or heat map etc.
# Analysis: you could check the data distributions from each attributes, the mean or std in statistical analysis, and the correlation in association.
```

```python
run_app(df, mode = 'external') # Display the plot in an externally launched window
```
